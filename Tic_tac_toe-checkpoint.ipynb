# Tic-Tac-Toe Game
In this project, I have created a game of Tic-Tac-Toe using tkinter GUI in python. I have created a 2 frame application. The first frame welcomes the players to this game and asks the names of the players. When the players submit their details, the frame containing the actual game gets displayed. The players can click a button from the available grid to play their turn. Once the game gets over, they can restart the game which redirects the players to the welcome page.  

## Importing Libraries

```python
import tkinter as tk
from tkinter import ttk
```

## Important Functions
Here, I will define the important functions necessary for successfully running the Tic-Tac-Toe app.

```python
#Declaring global variables
cur_player = 'X'
game_on = True
winner = '-'

#Defining functions
def check_row():
    global game_on
    global winner
    if user1.get() == 'X' and user2.get() == 'X' and user3.get() == 'X':
        game_on = False
        winner = 'X'
    elif user4.get() == 'X' and user5.get() == 'X' and user6.get() == 'X':
        game_on = False
        winner = 'X'
    elif user7.get() == 'X' and user8.get() == 'X' and user9.get() == 'X':
        game_on = False
        winner = 'X'
    elif user1.get() == 'O' and user2.get() == 'O' and user3.get() == 'O':
        game_on = False
        winner = 'O'
    elif user4.get() == 'O' and user5.get() == 'O' and user6.get() == 'O':
        game_on = False
        winner = 'O'
    elif user7.get() == 'O' and user8.get() == 'O' and user9.get() == 'O':
        game_on = False
        winner = 'O'

def check_column():
    global game_on
    global winner
    if user1.get() == 'X' and user4.get() == 'X' and user7.get() == 'X':
        game_on = False
        winner = 'X'
    elif user2.get() == 'X' and user5.get() == 'X' and user8.get() == 'X':
        game_on = False
        winner = 'X'
    elif user3.get() == 'X' and user6.get() == 'X' and user9.get() == 'X':
        game_on = False
        winner = 'X'
    elif user1.get() == 'O' and user4.get() == 'O' and user7.get() == 'O':
        game_on = False
        winner = 'O'
    elif user2.get() == 'O' and user5.get() == 'O' and user8.get() == 'O':
        game_on = False
        winner = 'O'
    elif user3.get() == 'O' and user6.get() == 'O' and user9.get() == 'O':
        game_on = False
        winner = 'O'

def check_diagonal():
    global game_on
    global winner
    if user1.get() == 'X' and user5.get() == 'X' and user9.get() == 'X':
        game_on = False
        winner = 'X'
    elif user3.get() == 'X' and user5.get() == 'X' and user7.get() == 'X':
        game_on = False
        winner = 'X'
    elif user1.get() == 'O' and user5.get() == 'O' and user9.get() == 'O':
        game_on = False
        winner = 'O'
    elif user3.get() == 'O' and user5.get() == 'O' and user7.get() == 'O':
        game_on = False
        winner = 'O'

#To check if there are any more possible moves
def check_possible():
    global game_on
    global winner
    if user1.get() != '-' and user2.get() != '-' and user3.get() != '-' and user4.get() != '-' and user5.get() != '-' and user6.get() != '-' and user7.get() != '-' and user8.get() != '-' and user9.get() != '-':
        game_on = False
        winner = 'none'

def check_winner():
    global game_on
    global winner
    check_row()
    check_column()
    check_diagonal()
    check_possible()
    if game_on == False and winner != 'none':
        if winner == 'O':
            this.set(f"Congratulations! {player2.get()} has won the game.")
        else: 
            this.set(f"Congratulations! {player1.get()} has won the game.")
    elif game_on == False and winner == 'none':
        this.set("Tie!")
        
def click(text):
    global cur_player
    if text.get() == '-':
        text.set(cur_player)
        if cur_player == 'X':
            cur_player = 'O'
            this.set(f"{player2.get()}'s turn: ")
        else:
            cur_player = 'X'
            this.set(f"{player1.get()}'s turn: ")
        check_winner()
    else:
        pass

def res(frame):
    frame.tkraise()
    
def raise_frame(frame):
    global game_on
    global winner
    global cur_player
    user1.set('-')
    user2.set('-')
    user3.set('-')
    user4.set('-')
    user5.set('-')
    user6.set('-')
    user7.set('-')
    user8.set('-')
    user9.set('-')
    game_on = True
    winner = '-'
    cur_player = 'X'
    this.set(f"{player1.get()}'s turn: ")
    frame.tkraise()
```

## Setting Up the Window

```python
try:
    from ctypes import windll  
    windll.shcore.SetProcessDPIAwareness(1)
except:
    pass

root = tk.Tk()
root.title("Tic-Tac-Toe (2 Player)")

#Declaring variables to store user input
player1 = tk.StringVar()
player2 = tk.StringVar()
user1 = tk.StringVar(value = '-')
user2 = tk.StringVar(value = '-')
user3 = tk.StringVar(value = '-')
user4 = tk.StringVar(value = '-')
user5 = tk.StringVar(value = '-')
user6 = tk.StringVar(value = '-')
user7 = tk.StringVar(value = '-')
user8 = tk.StringVar(value = '-')
user9 = tk.StringVar(value = '-')
this = tk.StringVar()

"""Creating 2 different frames
start: welcome page to the game
main: page for playing the game 
"""
main = ttk.Frame(root, padding = (30, 10))
start = ttk.Frame(root, padding = (30, 10))

for frame in (main, start):
    frame.grid(row = 0, column = 0, sticky = 'ew')
```

## Welcome Frame
This is the frame which welcomes the users to the game of Tic-Tac-Toe and asks the users to enter their respective names.

```python
#Declaring the widgets
label_start1 = tk.Label(start, text = 'Welcome to the game of Tic-Tac-Toe')
label_start1.configure(font = ("helvetica", 15))
label_start2 = tk.Label(start, text = 'Player 1: ')
entry_start1 = tk.Entry(start, textvariable = player1, width = 15)
label_start3 = tk.Label(start, text = 'Player 2: ')
entry_start2 = tk.Entry(start, textvariable = player2, width = 15)
button_start = tk.Button(start, text = 'Start', command = lambda: raise_frame(main))

#Displaying the widgets
label_start1.grid(row = 0, column = 0, columnspan = 2, sticky = 'ew', padx = 5, pady = 5)
label_start2.grid(row = 1, column = 0, sticky = 'w', padx = 5, pady = 5)
entry_start1.grid(row = 1, column = 1, sticky = 'ew', padx = 5, pady = 5)
label_start3.grid(row = 2, column = 0, sticky = 'w', padx = 5, pady = 5)
entry_start2.grid(row = 2, column = 1, sticky = 'ew', padx = 5, pady = 5)
button_start.grid(row = 3, column = 0, columnspan = 2, sticky = "ew", padx = 5, pady = 5)
```

## Game Frame
This is the frame where the Tic-Tac-Toe game is actually played.

```python
#Declaring the widgets
displaytop = ttk.Label(main, textvariable = this)
displaytop.configure(font = ("helvetica", 12))
displaytop.grid(row = 0, column = 0, columnspan = 3, sticky = 'w', padx = 5, pady = 5)
button1 = ttk.Button(main, textvariable = user1, command = lambda: click(user1))
button2 = ttk.Button(main, textvariable = user2, command = lambda: click(user2))
button3 = ttk.Button(main, textvariable = user3, command = lambda: click(user3))
button4 = ttk.Button(main, textvariable = user4, command = lambda: click(user4))
button5 = ttk.Button(main, textvariable = user5, command = lambda: click(user5))
button6 = ttk.Button(main, textvariable = user6, command = lambda: click(user6))
button7 = ttk.Button(main, textvariable = user7, command = lambda: click(user7))
button8 = ttk.Button(main, textvariable = user8, command = lambda: click(user8))
button9 = ttk.Button(main, textvariable = user9, command = lambda: click(user9))
restart = ttk.Button(main, text = 'Restart', command = lambda: res(start))

#Displaying the widgets
button1.grid(row = 1, column = 0, sticky = 'ew')
button2.grid(row = 1, column = 1, sticky = 'ew')
button3.grid(row = 1, column = 2, sticky = 'ew')
button4.grid(row = 2, column = 0, sticky = 'ew')
button5.grid(row = 2, column = 1, sticky = 'ew')
button6.grid(row = 2, column = 2, sticky = 'ew')
button7.grid(row = 3, column = 0, sticky = 'ew')
button8.grid(row = 3, column = 1, sticky = 'ew')
button9.grid(row = 3, column = 2, sticky = 'ew')
restart.grid(row = 6, column = 0, columnspan = 3, sticky = 'ew', padx = 5, pady = 5)

root.mainloop()
```
